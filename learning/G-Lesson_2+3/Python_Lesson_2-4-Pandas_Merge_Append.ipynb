## Concatenation

Concatenation basically glues together DataFrames.

axis = 0 >> attach rows, like union in SQL

axis = 1 >> attach columns, like join in SQL

<img height='200' src='arr_shape.png' width='500' title='Axis Explanation'>

```python
week1.head(3)
```

```python
week2.head(3)
```

#### Generating new index

```python
# Concat the two df
# axis = 0 >> attach rows
# axis = 1 >> attach columns
# >> use parameter: ignore_index=True >> generate new index

pd.concat ([week1, week2])
```

#### Using multi index

```python
# concat df, and use multi index
sales = pd.concat ([week1, week2], keys=['Week_1', 'Week_2'])
sales
```

```python
# locate rows with multi index
# syntax: df.loc [(external_index, internal_index), column_label]

sales.loc['Week_1']
# sales.loc[('Week_1', 3),'Customer ID']
```

## Append Method

The append() function is used to append rows of other to the end of caller, returning a new object.

Columns in other that are not in the caller are added as new columns

```python
# Append df at the end of the first df lines
# >> Can use parameter ignore_index = True
week2.append(week1, ignore_index=True)
```

## Merging

The merge () method allows you to merge DataFrames together using a similar logic as merging SQL Tables together

<img height='800' src='joins_type.png' width='1000' title='Join Types'>

#### Inner join

```python
# Customer ID 155 has 2 rows - why?
# >> can use parameter: suffixes = [' - A', ' - B']

week1.merge(week2, how = 'inner', on = 'Customer ID')

# week1.merge(week2, how = 'inner', on = 'Customer ID', suffixes = [' - A', ' - B'])
```

```python
week1[week1['Customer ID'] == 155]
```

```python
week2[week2['Customer ID'] == 155]
```

```python
# Merging based on two fields - shows only the rows that Customer ID and Food ID is same
week1.merge(week2, how = 'inner', on = ['Customer ID', 'Food ID'])
```

```python
week1[week1['Customer ID'] == 21]
```

```python
week2[week2['Customer ID'] == 21]
```

#### Outer joins

```python
# Show rows from both df
# Because there are NaN values - the data type converted to float
# >> Can use parameter: indicator = True
merged = week1.merge(week2, how = 'outer', on = 'Customer ID', suffixes = [' - Week 1', ' - Week 2'],
            indicator = True)
merged
```

```python
# To see how many data from each df
merged['_merge'].value_counts()
```

```python
# Cut off the rows from both df
# isin ([list_of_values_to _check])
mask = merged['_merge'].isin(['left_only', 'right_only'])
merged[mask]
```

#### Left joins

```python
# Combine the two df into week1 df
# >> Can use parameter: sort = True

week1 = week1.merge(foods, how = 'left', on = 'Food ID')
# week1 = week1.merge(foods, how = 'left', on = 'Food ID', sort = True)
```

```python
week1
```

```python
week1.head(10)
```

#### Merge based on different column names

When we don't have same column names - we will use left_on and right_on parameters

```python
customers.head(3)
```

```python
# When we don't have same column names
# Duplicate column ID, Customer ID
# >> Can use Method: drop('ID', axis = 'columns')

week2 = week2.merge(customers, how = 'left', left_on = 'Customer ID', right_on = 'ID')
# week2 = week2.merge(customers, how = 'left', left_on = 'Customer ID', right_on = 'ID', sort = True).drop('ID', axis = 'columns')
```

```python
week2.head()
```

#### Merge based on index

When the df index is its identifier

```python
# Re import customers and use ID as index
customers = pd.read_csv('Restaurant - Customers.csv', index_col='ID')
customers.head()
```

```python
# Merge df when in one df the matching column is the index
# we will use right_index = True
sales = week1.merge(customers, how = 'left', left_on = 'Customer ID', right_index = True)
```

## Stack () Method

The stack() function is used to stack the prescribed level(s) from columns to index. Return a reshaped DataFrame or Series having a multi-level index
