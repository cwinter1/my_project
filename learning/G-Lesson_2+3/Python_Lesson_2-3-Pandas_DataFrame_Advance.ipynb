```python
import numpy as np
import pandas as pd
```

## Generating random DataFrame

```python
# Generating random matrix
np.random.randint(1, 11, (5,4))
```

```python
# Cast matrix into DataFrame 
# Parameters:  data = random
#              index = list of the index labels
#              columns = list of the column names

df = pd.DataFrame(np.random.randint(1, 11, (5,4)), 
                  index=['A', 'B' ,'C', 'D', 'E'], 
                  columns=['W' ,'X', 'Y', 'Z'])
df
```

## Locate rows and columns

loc[start_row:stop:step, start column:stop:step]  = location = the index label >> include last column

iloc[start_row:stop:step, start column:stop:step] = integer location = row number starts with 0 >> not incluse last column

#### Locate row

```python
# Locate single row
df.loc['B']

# Locate range of rows
# df.loc['B':'D']

# locate single row
# df.iloc[1]
```

#### Locate column

```python
# Locate single column
df.loc[:, 'X']

# Locate range of columns
# df.loc[:, 'W':'Y']

# Locate range of columns (not include )
# df.iloc[:,0:3]

# df['X']
# df[   ['X', 'Y', 'W']   ]
```

#### Locate row and column

```python
# Locate list of rows and columns loc[   [list_of_rows] , [list_of_columns]   ]
df.loc[['A','D'],['W','Z']]
```

```python
# Locate range of rows and columns loc[   [start_row:stop:step] , [start_column:stop:step]   ]
df.loc['A':'D','W':'Y']
```

```python
# Update values ti Null values
df.loc['B','X'] = None
df.loc['D','Y'] = None
```

### Conditional Selection

An important feature of pandas is conditional selection using bracket notation:

```python
# Select rows that the value in column W, is bigger than 3
df[   df['W']>3   ]

# Select column x of the rows that the value un column W is bigger than 3
# df[ df['W']>3 ]['X']
```

```python
# Define masks variables with the conditions, and them choose df[mask1 | mask2 & mask 3]...
mask = df['W']>3
df[mask]
```

```python
# Combine 2 conditions with and operator
# df[ (df['Y'] > 3) & (df['W'] > 1) ]

mask1 = df['Y'] > 3
mask2 = df['W'] > 1

df[mask1 & mask2]
```
