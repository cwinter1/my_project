## Import libraries

```python
# import numpy as np
# import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Show plots inside jupyter notebook
%matplotlib inline
```

### Load build-in datasets

```python
# Load a built in dataset based on US State car crash percentages
crash_df = sns.load_dataset('car_crashes')
tips_df = sns.load_dataset('tips')
```

## Styling
You can set styling for your axes and grids

#### Grid style

```python
# >> Optional values: white, darkgrid, whitegrid, dark, ticks
sns.set_style('white')

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

#### Context style

```python
# Change size of lables, lines and other elements to best fit of how you will present your data
# >> Optional values: notebook (default), paper, talk, poster
# font_scale=  size of the font
sns.set_context('paper', font_scale=1.0)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

#### Remove spines

```python
# Get rid of spines
# must be written after the plot command
# Optional parameters: right=True, left=True, bottom=True, top=True

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
sns.despine(left=True)
# sns.despine(left=True, bottom=True)
```

#### Color palette
You can use Matplotlibs color maps for color styling

https://matplotlib.org/3.3.1/tutorials/colors/colormaps.html

```python
# Check the palette colors 
sns.palplot (sns.color_palette('nipy_spectral_r',15))
# sns.palplot (sns.color_palette('magma',15))
# sns.palplot (sns.color_palette('GnBu',15))
# sns.palplot (sns.color_palette('Pastel1_r',15))
```

```python
# Tip: to see the available palette list, write a wrong palette name, and check the error massage
sns.set_palette ('BrBG_r')
sns.pairplot(tips_df, hue='sex')
```

#### Customize x and y labels
Customizign the plot after it is created

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')

# Axis labels
plt.ylabel ('Y Label')
plt.xlabel ('X Label')

# Plot title
plt.title ('Plot title')

# Ticks mark rotation
plt.xticks (rotation=45)
plt.yticks (rotation=45)

# Axis value ranges
plt.xlim (2,8)
plt.ylim (2,8)
```

#### Change location of legend
Moves legend (after we create the plot)

https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.legend.html

```python
# 0=best, 1=upper right, 2=upper left, 3=lower left, 4=lower right ... 10=center

sns.boxplot(x='day',y='total_bill',data=tips_df, hue='sex')
plt.legend(loc=10)
```

```python
# You can supply a tuple of x & y from lower left
# Use parameter: bbox_to_anchor=(x, y)
#                title='Legend_title'
#                edgecolor='color' = change the color of legend edge 

sns.boxplot(x='day',y='total_bill',data=tips_df, hue='sex')

plt.legend(bbox_to_anchor=(1.2, 1))
# plt.legend(loc='best', bbox_to_anchor=(1.2,1), title='Sex')
# plt.legend(loc='best', bbox_to_anchor=(1.2,1), title='Sex', edgecolor='red')
```

#### Change the size of plot

```python
plt.figure(figsize=(14,9))
sns.boxplot(x='day',y='total_bill',data=tips_df, hue='sex')
```

```python
# Pre define the chart parameters:
# ylabel = label of y axis
# xlabel = label of x axis
# title = chart's title
# tick_params = controlling the ticks appearance
# xlim = the x axis ticks range

b_plot = sns.boxplot(x='day',y='total_bill',data=tips_df, hue='sex')

plt.xlabel('Day of week', fontweight='bold', color = 'green', fontsize='12', horizontalalignment='center')
plt.ylabel('Total bill Amount', fontweight='bold', color = 'green', fontsize='12')
plt.title ('Spend money in restaurant per day', fontsize=17, color='blue')

plt.tick_params(axis='both', colors='green', direction='in', length=12, width=2)
```
