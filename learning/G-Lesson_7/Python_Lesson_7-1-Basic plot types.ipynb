## Seaborn

* Seaborn is a data visualization library built on top of Matplotlib. 
* It is often used because it makes attractive visualizations and works easily with Pandas. 
* While in Matplotlib you often had to write multiple lines of code to create a plot Seaborn makes assumptions on what you want which often translates into getting the same plot with 1 line of code.

### Joint Plot
Jointplot compares 2 distributions and plots a scatter plot by default

## Categorical Plots

### Bar plots
A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars

```python
# Aggregate categorical data based on a function (mean is the default)
# A 95% confidence interval is a range of values that you can be 95% certain contains the true mean of the population

# >> Use parameter: estimator=np.std  = Statistical function to estimate within each categorical bin
# provided by NumPy : median, std, var, cov or make your own functions

# Estimate total bill amount based on sex

sns.barplot(x='sex',y='total_bill',data=tips_df)
# sns.barplot(x='sex',y='total_bill',data=tips_df, estimator=np.median)
# sns.barplot(x='sex',y='total_bill',data=tips_df, estimator=np.std)
# sns.barplot(x='sex',y='total_bill',data=tips_df, estimator=np.cov, color='red')
```

```python
# Same values as in pivot table- for comparison
tips_df.pivot_table(values='total_bill',columns='sex', aggfunc='mean')
```

## Count Plot
count plot is like a bar plot, but the estimator is counting the number of occurances

```python
# displays the number of items in each distinct value
sns.countplot(x='sex',data=tips_df)
```

## Box Plot
A box plot allows you to compare different variables

```python
# The box shows the quartiles of the data. The bar in the middle is the median and
# the box extends 1 standard deviation from the median
# The whiskers extend to all the other data aside from the points that are considered to be outliers
# >> Use parameter:hue='col_name' = can add another category being sex


# We see men spend way more on Friday versus less than women on Saturday
sns.boxplot(x='day',y='total_bill',data=tips_df)
# sns.boxplot(x='day',y='total_bill',data=tips_df, hue='sex')
```

```python
# Change styling
plt.figure(figsize=(14,9))
# sns.set_style('darkgrid')

# Create the plot
sns.boxplot(x='day',y='total_bill',data=tips_df, hue='sex')

# Moves legend:
# 0=best position, 1=right top corner, 2=left top corner
# 3=left bottom corner, 4=right bottom corner
plt.legend(loc=0)
```

## Violin Plot
Violin Plot is a combination of the boxplot and KDE

```python
# While a box plot corresponds to data points, the violin plot uses the KDE estimation of the data points
# >> >> Use parameter:split=True = allows you to compare how the categories compare to each other

sns.violinplot(x='day',y='total_bill',data=tips_df, hue='sex')
# sns.violinplot(x='day',y='total_bill',data=tips_df, hue='sex',split=True)
```

## Strip Plot
Draws a scatter plot representing all data points where one variable is categorical

```python
plt.figure(figsize=(8,5))

# It is often used to show all observations with a box plot that represents the average distribution
# >> Use parameter: jitter=True = spreads data points out so that they aren't stacked on top of each other
#                   hue='col_name' = breaks data into men and women
#                   dodge=True = separates the men and women data

sns.stripplot(x='day',y='total_bill',data=tips_df)
# sns.stripplot(x='day',y='total_bill',data=tips_df, jitter=True)
# sns.stripplot(x='day',y='total_bill',data=tips_df, jitter=True, hue='sex')
# sns.stripplot(x='day',y='total_bill',data=tips_df, jitter=True, hue='sex', dodge=True)
```

## Swarm Plot
Swarm plot is like a strip plot, but points are adjusted so they don't overlap

```python
# It looks like a combination of the violin and strip plots
sns.swarmplot(x='day',y='total_bill',data=tips_df)
```

```python
# You can stack a violin plot with a swarm
sns.violinplot(x='day',y='total_bill',data=tips_df)
sns.swarmplot(x='day',y='total_bill',data=tips_df, color='white')
```

## PairGrid

```python
# Dataset about iris flowers:
iris = sns.load_dataset('iris')
iris
```

#### Create empty grid

```python
# You can create a grid of different plots with complete control over what is displayed
# Create the empty grid system using the provided data
# Colorize based on species
iris_g = sns.PairGrid(iris, hue='species')
```

#### Same plot on all grid

```python
# You can define define variables for x & y for a custom grid
iris_g = sns.PairGrid(iris, hue='species',
                      x_vars=['sepal_length', 'sepal_width'],
                      y_vars=['petal_length', 'petal_width'])

# Put a scatter plot across the upper, lower and diagonal
iris_g.map(plt.scatter)

# Add a legend last
iris_g.add_legend()
```

#### Customize the plots in the grid

```python
# plt.figure(figsize=(8,6))
# sns.set_context('paper', font_scale=1.4)

iris_g = sns.PairGrid(iris, hue='species')

# Put a histogram on the diagonal 
iris_g.map_diag(plt.hist)

# Put a scatter plot every place else 
# iris_g.map_offdiag(plt.scatter)

# Add a legend last
iris_g.add_legend()
```

```python
iris_g = sns.PairGrid(iris, hue='species')

# Have different plots in upper, lower and diagonal
iris_g.map_diag(plt.hist)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)

# Add a legend last
iris_g.add_legend()
```
