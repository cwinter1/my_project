══════════════════════════════════════════════════════════════
 CONCEPT 6 — PYTEST: THE PRODUCTION TEST RUNNER
══════════════════════════════════════════════════════════════
 ROYAL ROAD STANDARD — W7
 The test functions in Task 4 used manual try/except — a good start.
 pytest is the industry standard: it discovers test_ functions automatically,
 runs them in isolation, and reports which passed or failed with a clear diff.

 In production, test files are saved as test_pipeline.py and run with:
   pytest test_pipeline.py -v

 pytest rules:
   - Function names must start with test_
   - Use plain assert — no try/except needed
   - pytest catches AssertionError and formats the failure for you
   - One test = one behaviour; keep tests small and focused

 Since we are in a notebook, we use a run_tests() helper to simulate pytest.
 The pattern is identical — only the runner is different.

 EXAMPLE ──────────────────────────────────────────────────────

```python
# pytest pattern demonstrated inline
# In production this lives in test_pipeline.py and you run: pytest test_pipeline.py -v

def add(a, b):
    return a + b

def multiply(a, b):
    return a * b


# Test functions: start with test_, use assert directly
def test_add_basic():
    assert add(2, 3) == 5

def test_add_zero():
    assert add(0, 0) == 0

def test_multiply():
    assert multiply(3, 4) == 12

def test_multiply_negative():
    assert multiply(-2, 5) == -10


# Notebook runner that mimics pytest -v output
def run_tests(*fns):
    passed, failed = 0, 0
    for fn in fns:
        try:
            fn()
            print(f"PASS  {fn.__name__}")
            passed += 1
        except AssertionError as e:
            print(f"FAIL  {fn.__name__}: {e}")
            failed += 1
    print(f"\n{passed} passed, {failed} failed")


run_tests(test_add_basic, test_add_zero, test_multiply, test_multiply_negative)
```
