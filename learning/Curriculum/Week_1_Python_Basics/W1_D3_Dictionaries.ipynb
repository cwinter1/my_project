 CONCEPT 1 — CREATING A DICTIONARY
══════════════════════════════════════════════════════════════

 A dictionary stores data as key-value pairs, enclosed in curly braces {}.
 Each pair is written as:  "key": value
 Pairs are separated by commas.

 Syntax:
   my_dict = {"key1": value1, "key2": value2}

 Keys are usually strings. Values can be any type.

 Dictionary vs List:
   List:  ordered by position — you access items by index number (0, 1, 2...)
   Dict:  labeled by name — you access items by key ("name", "price"...)

 Use a dict when each value has a meaningful label (like a row of data).

EXAMPLE ──────────────────────────────────────────────────────


```python

employee = {
    "name":       "Avi Ben-David",
    "age":        41,
    "department": "Data Engineering",
    "salary":     110000.0,
    "is_active":  True
}

print(employee)

```

 CONCEPT 2 — ACCESSING VALUES
══════════════════════════════════════════════════════════════

 Two ways to get a value from a dictionary:

 1. Bracket notation:  dict["key"]
    - Returns the value for that key
    - Raises a KeyError if the key does not exist

 2. .get() method:  dict.get("key")
    - Returns the value if the key exists
    - Returns None if the key does not exist (no crash)
    - You can provide a fallback:  dict.get("key", "default value")

 Use .get() whenever the key might not be present — it is the safe choice.

EXAMPLE ──────────────────────────────────────────────────────


```python

print(employee["name"])                            # Avi Ben-David
print(employee["salary"])                          # 110000.0
print(employee.get("department"))                  # Data Engineering
print(employee.get("bonus"))                       # None — key does not exist
print(employee.get("bonus", "No bonus data"))      # No bonus data — fallback used

```

 CONCEPT 3 — ADD, UPDATE, DELETE
══════════════════════════════════════════════════════════════

 Dictionaries are mutable — you can change them after creation.

 Add a new key:     dict["new_key"] = value
 Update a key:      dict["existing_key"] = new_value   (same syntax as add)
 Delete a key:      del dict["key"]

 Python uses the same syntax for add and update.
 If the key exists, the value is replaced. If it does not exist, it is created.

EXAMPLE ──────────────────────────────────────────────────────


```python

employee["department"] = "Platform Engineering"   # update existing key
print(employee)

employee["years_at_company"] = 5                  # add new key
print(employee)

del employee["is_active"]                         # delete a key
print(employee)

```

 CONCEPT 4 — LOOPING WITH .items()
══════════════════════════════════════════════════════════════

 .items() lets you loop through a dictionary and access both
 the key and the value at the same time.

 Syntax:
   for key, value in my_dict.items():
       print(key, value)

 You can also list all keys or all values:
   my_dict.keys()    — returns all keys
   my_dict.values()  — returns all values

 Looping with .items() is the most common way to process dict data row by row.

EXAMPLE ──────────────────────────────────────────────────────


```python

monthly_revenue = {"January": 85000, "February": 92000, "March": 78000, "April": 105000}

for month, revenue in monthly_revenue.items():
    print(month + ":", revenue)

print("All months:", list(monthly_revenue.keys()))
print("All revenues:", list(monthly_revenue.values()))

```

 CONCEPT 5 — DATACLASSES: STRUCTURED DATA CONTRACTS
══════════════════════════════════════════════════════════════
 ROYAL ROAD STANDARD — W1
 Every data shape in your code should be a dataclass (or Pydantic model later).
 Never pass raw dicts between functions when you know the structure in advance.

 A dataclass is a regular class where Python auto-generates __init__, __repr__,
 and __eq__ based on the field declarations you write.

 Why this matters in production:
   - Your data has a documented, enforced shape
   - You get type hints that IDEs and linters can check
   - Comparisons and printing work automatically
   - You catch a missing field at object creation, not 10 functions later

 Syntax:
   from dataclasses import dataclass, field

   @dataclass
   class MyRecord:
       name: str
       value: float
       tags: list = field(default_factory=list)  # mutable default — use field()

 EXAMPLE ──────────────────────────────────────────────────────

```python
from dataclasses import dataclass, field


@dataclass
class Employee:
    name: str
    department: str
    salary: float
    is_active: bool = True
    skills: list = field(default_factory=list)

    def annual_bonus(self, rate: float = 0.10) -> float:
        if not self.is_active:
            return 0.0
        return round(self.salary * rate, 2)


emp1 = Employee("Alice Ng", "Engineering", 95000, skills=["Python", "SQL"])
emp2 = Employee("Bob Chen", "Sales", 72000)

print(emp1)
# Employee(name='Alice Ng', department='Engineering', salary=95000, is_active=True, skills=['Python', 'SQL'])

print(f"Bonus: ${emp1.annual_bonus()}")   # Bonus: $9500.0
print(f"Inactive bonus: ${emp2.annual_bonus(emp2.is_active)}")

# __eq__ is generated automatically
emp3 = Employee("Alice Ng", "Engineering", 95000, skills=["Python", "SQL"])
print(f"emp1 == emp3: {emp1 == emp3}")   # True
```

 CONCEPT 6 — TESTING YOUR DATA CONTRACT
══════════════════════════════════════════════════════════════

 Shift-Left Testing: verify your data contract right where you define it.
 You do not need a test framework yet. A plain assert statement raises
 AssertionError if the condition is False — that is enough to catch contract
 violations early, while the code is still in front of you.

 Syntax:
   assert condition, "Message shown if the condition is False"

 What to test on a dataclass:
   - Field types match the declared type  (isinstance)
   - Default values are correct
   - Computed methods return the right result for known inputs

 The rule: if you define a contract, write at least one assert per field
 and one assert per method. This runs in the notebook as you work.
 No separate test file yet — that comes in Week 7.

EXAMPLE ──────────────────────────────────────────────────────

```python
# Employee is defined in the cell above. This cell asserts its contract holds.
emp = Employee("Sara Levi", "Engineering", 90000.0)

assert isinstance(emp.name, str),         "name must be a string"
assert isinstance(emp.department, str),   "department must be a string"
assert isinstance(emp.salary, float),     "salary must be a float"
assert isinstance(emp.is_active, bool),   "is_active must be a bool"
assert emp.is_active is True,             "default is_active must be True"
assert emp.skills == [],                  "default skills must be an empty list"

assert emp.annual_bonus() == 9000.0,      "10% of 90000.0 must equal 9000.0"
assert emp.annual_bonus(0.05) == 4500.0,  "5% of 90000.0 must equal 4500.0"

inactive = Employee("Ron Bar", "Ops", 60000.0, is_active=False)
assert inactive.annual_bonus() == 0.0,    "inactive employee must receive 0.0 bonus"

print("Employee contract: all assertions passed.")
```
