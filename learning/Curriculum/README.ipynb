HOW THIS CURRICULUM WORKS:
---------------------------
Each lesson is a .py file you open and run in VS Code.
Every lesson is structured as: CONCEPT → EXAMPLE → EXERCISE (repeated per topic)

  CONCEPT  → plain English explanation as comments (just read)
  EXAMPLE  → working code that runs — read it, then run the file
  EXERCISE → your turn: write code below the "# YOUR CODE HERE ↓" marker
             Check: does your output match "# Expected output:" in the comments?

TIME PER LESSON: ~30-35 minutes  (3 concepts × 10 min each)
PACE: 1 lesson per day (5 days/week = 11 weeks total)

HOW TO RUN A LESSON:
---------------------
  1. Open the lesson file in VS Code
  2. Read the first CONCEPT section (comments only)
  3. Read the EXAMPLE code, then press Run → see the output
  4. Write your answer in Exercise 1, run again
  5. Continue: read Concept 2 → Example 2 → Exercise 2 → etc.
