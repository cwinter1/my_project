 CONCEPT 1 — WHAT IS AN LLM API?

══════════════════════════════════════════════════════════════

LLM = Large Language Model (GPT-4, Claude, Gemini, Llama, etc.)
API = a way for your Python code to send a message to the model and get a reply.

Think of it like texting a very smart assistant:
  You send:  "Summarize this text: [your text]"
  It replies: "Here is a summary..."

The two most important concepts:

SYSTEM PROMPT  — instructions given to the model before the conversation starts
                 "You are a data analyst. Always respond with concise JSON."

USER MESSAGE   — what the user (your code) sends
                 "Classify this customer review as positive, negative, or neutral."

TOKENS         — LLMs measure text in tokens, not words.
                 ~1 token = ~0.75 words.  You pay per token used.
                 Keep prompts concise to reduce cost.

HOW THE CALL WORKS:

 CONCEPT 5 — STREAMING RESPONSES
══════════════════════════════════════════════════════════════

By default, the LLM API waits until the full response is ready before sending it.
For a 200-word answer, that might be 5-15 seconds of silence.

STREAMING sends each token to your code as it is generated.
The user sees the answer appear word by word — like watching someone type.
This is how ChatGPT, Claude.ai, and every production AI interface works.

HOW IT WORKS:
  Normal call:   response = client.chat.completions.create(...)
  Streaming:     stream   = client.chat.completions.create(..., stream=True)

  With streaming, the response is an iterator.
  Each iteration gives you one chunk (a few tokens).
  You print each chunk immediately using print(token, end="", flush=True).

WHEN TO USE STREAMING:
  Use streaming:  user is waiting and watching a screen
  Use blocking:   you need the full text before processing it
                  (e.g., parsing JSON, calling another function with the result)

COST: streaming does not change the cost. You pay the same tokens either way.

EXAMPLE ──────────────────────────────────────────────────────

```python

print()
print("=" * 55)
print("CONCEPT 5: Streaming Responses")
print("=" * 55)

# Without streaming: your code waits for the FULL response, then displays it.
# This can take 5-30 seconds for long answers — the user sees nothing during that time.
#
# With streaming: tokens arrive one by one. The user sees text appear immediately.
# This is how ChatGPT, Claude.ai, and every real AI app works.

# OPENAI STREAMING:
# from openai import OpenAI
# client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))
#
# stream = client.chat.completions.create(
#     model="gpt-4o-mini",
#     messages=[
#         {"role": "system", "content": "You are a data analyst assistant."},
#         {"role": "user",   "content": "Explain what a data warehouse is in 3 sentences."}
#     ],
#     max_tokens=200,
#     stream=True            # <-- the only change from a normal call
# )
#
# full_text = ""
# for chunk in stream:
#     token = chunk.choices[0].delta.content
#     if token:
#         print(token, end="", flush=True)   # print each token as it arrives
#         full_text += token
# print()    # newline at the end
# print(f"\nTotal characters received: {len(full_text)}")

# ANTHROPIC STREAMING:
# import anthropic
# client = anthropic.Anthropic(api_key=os.environ.get("ANTHROPIC_API_KEY"))
#
# with client.messages.stream(
#     model="claude-haiku-4-5-20251001",
#     max_tokens=200,
#     system="You are a data analyst assistant.",
#     messages=[{"role": "user", "content": "Explain what a data warehouse is in 3 sentences."}]
# ) as stream:
#     for text in stream.text_stream:
#         print(text, end="", flush=True)
# print()

print("Streaming code shown above — uncomment once you have an API key.")
print()
print("Key difference:")
print("  Blocking:  call → wait → get full response")
print("  Streaming: call → get token 1 → get token 2 → ... → done")
print()
print("Use streaming whenever the user is waiting and watching a screen.")
print("Use blocking when you need the full text before doing something with it.")
```

 CONCEPT 6 — ASYNC / PARALLEL API CALLS
══════════════════════════════════════════════════════════════

When you need to process many items (100s of reviews, records, documents),
calling the API one at a time is too slow for production.

SEQUENTIAL (slow):
  for review in reviews:
      result = call_llm(review)   # waits 1-3 seconds before moving on

ASYNC / PARALLEL (fast):
  tasks = [call_llm(r) for r in reviews]
  results = await asyncio.gather(*tasks)   # all calls run at the same time

Python's asyncio lets you send multiple API calls simultaneously.
The openai and anthropic libraries both support async with AsyncOpenAI / AsyncAnthropic.

RATE LIMITS:
  APIs allow a maximum number of requests per minute (e.g., 500 RPM).
  If you fire 1000 requests at once, you'll hit the limit and get errors.
  Solution: asyncio.Semaphore(N) — allows at most N concurrent requests.

  semaphore = asyncio.Semaphore(5)   # max 5 at a time
  async with semaphore:
      result = await client.chat.completions.create(...)

WHEN TO USE ASYNC:
  Use async: batch jobs, processing many records, data pipelines
  Use sync:  single interactive calls, simple scripts, testing

INSTALL:
  pip install openai   (AsyncOpenAI is included in the same package)

EXAMPLE ──────────────────────────────────────────────────────

```python

import asyncio
import time

print()
print("=" * 55)
print("CONCEPT 6: Async / Parallel API Calls")
print("=" * 55)

# SEQUENTIAL vs ASYNC:
# Suppose each API call takes 1 second.
# Sequential: 1000 calls × 1s = ~17 minutes
# Async:      1000 calls in ~2-5 seconds (limited by rate limits, not waiting)

# ASYNC PATTERN WITH OPENAI:
# from openai import AsyncOpenAI
# import asyncio
#
# client = AsyncOpenAI(api_key=os.environ.get("OPENAI_API_KEY"))
# semaphore = asyncio.Semaphore(5)   # max 5 concurrent requests at once
#
# async def classify_one(text):
#     async with semaphore:
#         response = await client.chat.completions.create(
#             model="gpt-4o-mini",
#             messages=[
#                 {"role": "system", "content": "Classify as: positive / negative / neutral. One word only."},
#                 {"role": "user",   "content": text}
#             ],
#             max_tokens=5
#         )
#     return response.choices[0].message.content.strip().lower()
#
# async def classify_batch(texts):
#     tasks = [classify_one(t) for t in texts]
#     return await asyncio.gather(*tasks)   # run all tasks concurrently
#
# # Run the batch:
# reviews = ["Great product!", "Terrible quality", "It's okay"]
# results = asyncio.run(classify_batch(reviews))
# for review, label in zip(reviews, results):
#     print(f"  {review[:45]:45} -> {label}")

# Simulate the time difference:
def sequential_simulation(n_items, seconds_per_call=1.0):
    return n_items * seconds_per_call

def async_simulation(n_items, concurrency=5, seconds_per_call=1.0):
    batches = (n_items + concurrency - 1) // concurrency
    return batches * seconds_per_call

n = 50
print(f"\nProcessing {n} reviews:")
print(f"  Sequential: ~{sequential_simulation(n):.0f} seconds")
print(f"  Async (5 concurrent): ~{async_simulation(n):.0f} seconds")
print(f"  Speedup: ~{sequential_simulation(n) / async_simulation(n):.0f}x")

print()
print("Key points:")
print("  - Use asyncio.Semaphore to stay within API rate limits")
print("  - asyncio.gather() runs all tasks concurrently")
print("  - Use AsyncOpenAI (not OpenAI) for async calls")
print("  - Results from gather() are in the same order as input")
```

══════════════════════════════════════════════════════════════
 CONCEPT 7 — LLM SEMANTIC CACHING: SKIP REPEAT API CALLS
══════════════════════════════════════════════════════════════
 LLM API calls are expensive (cost) and slow (latency).
 If the same prompt is sent twice, the second call is wasted.

 LLM semantic caching:
   Key   : MD5 hash of (model + prompt) — deterministic fingerprint
   Store : {key: {"response": str, "cached_at": float, "cost_usd": float}}
   Hit   : return stored response instantly (no API call, $0 cost)
   Miss  : call the API, store the response, log the cost

 Cost impact example:
   1000 calls/day × $0.002/call = $2.00/day
   If 60% are cache hits → 400 actual calls × $0.002 = $0.80/day
   Saving: $1.20/day = $438/year on ONE endpoint alone

 When NOT to cache:
   - temperature > 0 (responses vary — same prompt gives different answers)
   - Personalized prompts that include user-specific data
   - Real-time context (current time, live prices embedded in prompt)

 EXAMPLE ──────────────────────────────────────────────────────

```python
import hashlib
import time


class PromptCache:
    """In-memory LLM response cache with cost tracking."""

    def __init__(self, cost_per_call_usd: float = 0.002):
        self._store: dict = {}
        self._cost_per_call = cost_per_call_usd
        self._hits = 0
        self._misses = 0

    def _key(self, model: str, prompt: str) -> str:
        return hashlib.md5(f"{model}:{prompt}".encode()).hexdigest()

    def get(self, model: str, prompt: str):
        k = self._key(model, prompt)
        if k in self._store:
            self._hits += 1
            return self._store[k]["response"]
        self._misses += 1
        return None

    def set(self, model: str, prompt: str, response: str) -> None:
        k = self._key(model, prompt)
        self._store[k] = {"response": response, "cached_at": time.time()}

    def summary(self) -> dict:
        total = self._hits + self._misses
        actual_cost = self._misses * self._cost_per_call
        potential_cost = total * self._cost_per_call
        saving = potential_cost - actual_cost
        return {
            "total_calls": total,
            "cache_hits": self._hits,
            "cache_misses": self._misses,
            "hit_rate": f"{self._hits / total * 100:.1f}%" if total else "0%",
            "actual_cost_usd": round(actual_cost, 4),
            "saved_usd": round(saving, 4),
        }


# Simulate an LLM call (replace with real openai call in production)
def fake_llm(model: str, prompt: str) -> str:
    time.sleep(0.05)  # simulate latency
    return f"[{model}] Answer to: {prompt[:30]}..."


cache = PromptCache(cost_per_call_usd=0.002)

prompts = [
    ("gpt-4o-mini", "Summarize the Q1 sales report in 3 bullet points."),
    ("gpt-4o-mini", "What are the top 5 products by revenue?"),
    ("gpt-4o-mini", "Summarize the Q1 sales report in 3 bullet points."),   # repeat
    ("gpt-4o-mini", "Summarize the Q1 sales report in 3 bullet points."),   # repeat
    ("gpt-4o-mini", "What are the top 5 products by revenue?"),              # repeat
]

for model, prompt in prompts:
    cached = cache.get(model, prompt)
    if cached:
        response = cached
        label = "HIT "
    else:
        response = fake_llm(model, prompt)
        cache.set(model, prompt, response)
        label = "MISS"
    print(f"  [{label}] {prompt[:45]}")

print()
for k, v in cache.summary().items():
    print(f"  {k}: {v}")
```

```python
import os
from dotenv import load_dotenv

load_dotenv()
GROQ_API_KEY = os.environ.get('GROQ_API_KEY', '')

# ── Interface comparison: OpenAI vs Groq ──────────────────────
print('=== LLM Interface Comparison ===')
print()
print('OpenAI (ChatOpenAI):')
print('  from langchain_openai import ChatOpenAI')
print("  llm = ChatOpenAI(model='gpt-4o-mini', api_key=OPENAI_API_KEY, temperature=0)")
print()
print('Groq (ChatGroq) -- drop-in replacement:')
print('  from langchain_groq import ChatGroq')
print("  llm = ChatGroq(model='llama-3.1-70b-versatile', api_key=GROQ_API_KEY, temperature=0)")
print()
print('Both support the same interface:')
print("  result = llm.invoke('Summarize this data...')")
print("  for chunk in llm.stream('Explain...'): print(chunk.content, end='')")
print()

if GROQ_API_KEY:
    from langchain_groq import ChatGroq
    from langchain_core.messages import HumanMessage
    llm = ChatGroq(model='llama-3.1-70b-versatile', api_key=GROQ_API_KEY, temperature=0)
    result = llm.invoke([HumanMessage(content='In one sentence: what is DuckDB?')])
    print('Live Groq response:')
    print(f'  {result.content}')
else:
    print('GROQ_API_KEY not set -- live call skipped.')
    print('Expected output when key is set:')
    print('  DuckDB is an in-process analytical SQL database optimized for fast file queries.')
```
