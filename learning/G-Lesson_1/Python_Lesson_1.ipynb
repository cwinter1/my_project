## Condition - if block

````python
# It is important to note that
# comparison is represented by ==
# Assignment is represented by =

x = 30
y = 20

if x == y:
    print ("x = y = ", x)
elif x > y:
    print ("x is greater than y")
else:
    print ("x is lower than y")
````

````python
# Palindrom - a word that its characters read the same backward as forward i.e level, kayak

word = "hello"
#"Level"

if word.upper() == word[::-1].upper():
    print ("Palindrom")
else:
    print ("not Palindrom")
````

## While loop
    As long as the condition is true, the loop keep on iterating

````python
# Each loop calles iteration
i = 0

while i < 6:
    print (i)
    i += 1
````

## For loop

````python
# range provide values within a given range.
# range (start, stop, step)

range(5)
````

````python
# From 0 to n-1

n = 5
for i in range(n):
    print (i)
````

````python
for letter in 'Data Analyst':
    print(letter)
````

#### Continue
    Skip over one or more loops

````python
# print even numbers

for i in range(5):
    if i % 2 != 0:
        continue
    print (i)
````

#### Break
    Exit the loop

````python
for i in range(5):
    if i == 3:
        break
    print (i)
````

````python
for i in range (30,10,-2):
    print (i)
````

## Tuple

A tuple is an (immutable) ordered list of values. A tuple is in many ways similar to a list; one of the most important differences is that tuples can be used as keys in dictionaries and as elements of sets, while lists cannot.

You can use tuples to present things that shouldn't be changed, such as days of the week, or dates on a calendar.

````python
# Defining a Tuple variable
my_tuple = ('one',2,'Three',9, 2, 2)
````

````python
type (my_tuple)
````

````python
# Tuple variable is immutable - cannot be changed
my_tuple[1] = 3
````

````python
# With the help of Tuple, it is possible to perform multiple placements in an easy and elegant way

w, x, y, z = (1,2,3,4)
````

````python
# Returns the amount of items in the Tuple
len (my_tuple)
````

````python
# Use index to get to an item
my_tuple[1]
````

````python
# Use value to get to the index
my_tuple.index ('Three')
````

````python
# Count how many times the item 2 appears in the Tuple
my_tuple.count (2)
````
