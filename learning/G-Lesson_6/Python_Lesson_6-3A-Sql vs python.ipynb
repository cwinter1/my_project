## Import Data

```python
# Pandas
df = pd.read_csv('Fert_data.csv')
```

```python
country_sub_reg = [
    ['country', 'subregion'],
    ['Kenya', 'East Africa'],
    ['Liberia', 'West Africa'],
    ['Mali', 'West Africa']
]
df_sr = pd.DataFrame(country_sub_reg[1:],columns=country_sub_reg[0])
df_sr
```

```python
data = [
    ['country', 'region', 'tfr', 'contraceptors'],    
    ['USA', 'North.Amer', 1.77, 20],
    ['UK', 'Europe', 1.79, 23],
    ['Bangladesh', 'Asia', 5.5, 40],
    ['Thailand', 'Asia', 2.3, 68],
    ['Mali', 'Africa', 6.80, 5]
]
df1 = pd.DataFrame(data[1:], columns=data[0])
```

```python
# SQL
# In sql can import from csv

tbl_name = 'fert_data'
py_mssql.drop_table_if_exist(tbl_name)
py_mssql.create_tbl_from_df(df,tbl_name)
py_mssql.insert_into_tbl_from_df(df,tbl_name) 
```

```python
# SQL 
# Create country_sub_region table (in sql can import from csv)
tbl_name = 'country_sub_region'
py_mssql.drop_table_if_exist(tbl_name)
py_mssql.create_tbl_from_df(df_sr,tbl_name)
py_mssql.insert_into_tbl_from_df(df_sr,tbl_name) 
```

## Where

```python
# Python
mask1 = df['tfr'] >= 5.5
mask2 = df['tfr'] <= 7
mask3 = df['region'] != 'Africa'

df[mask1 & mask2 & mask3]

# df[(df['tfr'] >= 5.5) & (df['tfr'] <= 7) & (df['region'] != 'Africa')]
```

## Insert

```python
# Python
# In Pandas, We can use tools such as concat() or append() to perform this task

data = {
    'country': 'Japan', 
    'region': 'Asia', 
    'tfr':1.73, 
    'contraceptors' : 73
}
df = df.append(data, ignore_index=True)
df.tail(3)
```

## Update

```python
# Python
# df.loc[row_indexer,col_indexer] = value

mask = df['country'] == 'Mauitius'
df[mask]

# df.loc[mask, 'contraceptors'] = 65
# df.loc[mask]
```

```python
df = df.loc[df['region'] != 'Asia']
df
```

## Union

```python
df_dupli = pd.concat([df1, df])
df_dupli
```

```python
# we will get duplicate rows 
mask = df_dupli['country'] == 'Mali'
df_dupli[mask]
```

```python
# We can drop duplicate records using drop_duplicates()
pd.concat([df1, df]).drop_duplicates()
```

## Group by

```python
# we have a groupby() function that helps us in summarizing data along a specific column
df.groupby('region').agg({'tfr': 'mean', 'contraceptors': 'size'}).round(2)
```
