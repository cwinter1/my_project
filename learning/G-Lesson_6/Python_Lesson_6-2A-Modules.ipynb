#### MSSql integration: build diabets table using python - solution

1. read the diabetes csv file (in this lesson folder)


2. filter only Outcome=1


3. build sql table from df (table 'diabetes')


4. read from new created table to new df:
    * only columns Glucose,BloodPressure, DiabetesPedigreeFunction, Age, Outcome	
    * only BloodPressure above 80
    
    
5. group by age, aggregate over blood pressure: mean, std, min, max
 

```python
import pandas as pd
import pyodbc
import sys

sys.path.append('Modules/')
from Py_MSSql import Py_MSSql
```

```python
# define the connection using Py_MSSql module
# !! Run the SQL script file name 'Create_db.sql' to create TutorialDB database !!

server = 'localhost\\SQLEXPRESS;Trusted_Connection=True;' #' Integrated Security=True;'
database = 'TutorialDB'

py_mssql = Py_MSSql(server=server,database=database)
```

```python
# 1
df = pd.read_csv('diabetes.csv')
df.head(3)
```

```python
# 2
mask = df['Outcome'] == 1
df = df[mask]
df.head(3)
```

```python
# 3
tbl_name = 'diabetes'

# Check if exist, and drop, so we can re-create the table
py_mssql.drop_table_if_exist(tbl_name)

# Create new table in same stucture as DateFrame variable
py_mssql.create_tbl_from_df (df,tbl_name)

# Insert date to the table we created in MSSQL, from out DataFrame 
py_mssql.insert_into_tbl_from_df (df,tbl_name) 
```

```python
# 4
sql_query = ''' select  Glucose,
                        BloodPressure, 
                        DiabetesPedigreeFunction,
                        Age, 
                        Outcome
                from diabetes 
                where BloodPressure > 80 '''


sdf = py_mssql.read_sql_query(sql_query)
sdf.head()
```

```python
# 5
new_df = sdf.groupby('Age').agg({'BloodPressure': ['mean', 'std', 'min', 'max']})
new_df
```
