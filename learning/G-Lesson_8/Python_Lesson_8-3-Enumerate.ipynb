## Enumerate
* Enumerate is a built-in function of Python
* It allows us to loop over something and have an automatic counter

#### Example

```python
# Loop over the list
# Counter = counts the iterations
# Value = get the value of the item

my_list = ['apple', 'banana', 'grapes', 'pear']
for counter, value in enumerate(my_list):
    print (counter, value)
```

```python
# Use parameter: number of the first iteration (default is 0)

my_list = ['apple', 'banana', 'grapes', 'pear']
for c, value in enumerate(my_list, 4):
    print(c, value)
```
