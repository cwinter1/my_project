#### 6. Replace null values in the column with the highest count of nulls, with the mean value of that column

```python
# get nulls count per column - Cabin has the highest count, BUT IT'S NOT A NUMBER SO IT DOESN'T HAVE MEAN!
# the relevant column: Age
df.isnull().sum()
```

```python
#df['Cabin'] = df['Cabin'].fillna(df['Cabin'].mean()) 
df['Age'] = df['Age'].fillna(df['Age'].mean()) 
df.head(10)
```

```python
# Option 1
df['Fare_rounded'] = df['Fare'].astype(int)

# Option 2
df['Fare_rounded'] = df['Fare'].round(0)
df
```

#### 8. Add new column named 'age_decade' which is defined as a decade category of the age column:
#### e.g.: 23 -> '20+', 46 -> '40+' etc.   
#### (before doing so drop all rows with null ages)

```python
#drop nulls of age - we don't have nulls since we replaced it with mean in previous ex
df.dropna(subset=['Age'],inplace=True)
df.info()
```

```python
def toDecade(age):
    for i in range(10):
        print(str(int(age/10) * 10)+'+'+str(i))
        #return str(int(age/10) * 10)+'+'+str(i)

# test function
toDecade(34)
```

```python
# 1st option with a function
#df['age_decade'] = df['Age'].apply(toDecade)

# 2nd option with a lambda
df['age_decade'] = df['Age'].apply(lambda age: str(int(age/10) * 10)+'+' )
df['age_decade']
```

```python
df[['Age','age_decade']].head(5)
```

#### 9A. Group by the new created column 'age_decade', aggragating: mean and count  over columns 'survived' & 'fare'
#### 9B. Group by column 'sex', aggragating: mean, count and std over column 'survived'

```python
df.groupby('age_decade').agg({'Survived':['count','mean'],
                              'Fare':['count','mean','std']})   
```

```python
df.groupby('Sex').agg({'Survived':['count','sum','mean','std']}) 
                              
```

#### 10. Create a new partial dataframe from titanic with females only (make it a copy). 
#### merge the original df as the strong side of a 'left join', the key will be passenger id, leave all original df columns and only passenger id column of new female df, rename the column.

```python
mask = df['Sex']=='female'
df_female = df[mask].copy()
```

```python
df_female['PassengerId_rkey'] = df_female['PassengerId']
df_female.head(3)
```

```python

df_mrg = pd.merge(df, df_female[['PassengerId_rkey']]
                  ,how='left'
                  ,left_on=['PassengerId']
                  ,right_on=['PassengerId_rkey']
                  ,indicator=True) #indicator: the source of the row (left or both)
df_mrg
```
