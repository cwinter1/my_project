## Grouping

#### Group object
When using group by() method, we Create categories that based on common values in certain column or multiple columns based on DataFrame

```python
# Create groupby object and store in in variable
g_violation = df.groupby('violation_raw')
```

```python
# The groupby object is DataFrameGroupBy type
type(g_violation)
```

```python
# Number of groups
len (g_violation)

# Can get the same result with the function nunique()
# df['violation_raw'].nunique()
```

```python
# How many rows in each group category
g_violation.size()
```

```python
# Show the first/last row in each group category
g_violation.first()
# g_violation.last()
```

```python
# Returns divtionary with key=group, value = list of all indexes of rows in the group
g_violation.groups
# g_violation.groups['Speeding']
```

```python
# Returns rows from this group category
g_violation.get_group('Speeding')
```

```python
# Returns computed max/min/mean/count/sum of values within each group

g_violation['driver_age'].max()
# g_violation['driver_age'].min()
# g_violation['driver_age'].mean()
# g_violation['violation'].count()
# g_violation['ticket_amount'].sum()
# g_violation[['driver_age','ticket_amount']].max()
# g_violation['driver_age'].describe()
```

```python
# Create group with main category and sub category

g_gender_violation = df.groupby(['driver_gender','violation_raw'])
```

```python
# Rturns computed max/min/mean/count/sum of values within each group - multiple columns

g_gender_violation[['driver_age','ticket_amount']].mean()
```

## Group Aggregations

#### agg() method
Using group by with agg() to enable multiple columns + multiple aggregations at once


```python
# Choose only some of the columns

g_violation[['driver_age', 'ticket_amount']]
```

```python
# Calculate the three aggregations ('size', 'sum', 'mean') for all the columns
g_violation[['driver_age', 'ticket_amount']].agg(['size', 'sum', 'mean'])
```

```python
# Calculate different aggregations on different columns

g_violation.agg({'driver_age' : ['min', 'mean', 'max'],
                 'ticket_amount' : ['mean', 'sum']})
```

```python
# Short writing example
df.groupby(['driver_gender','violation_raw']).agg({'driver_age' : ['min', 'mean', 'max']})

# df.groupby(['driver_gender','violation_raw']).agg({'driver_age' : ['min', 'mean', 'max']}).to_csv('driver gender violation.csv')
```

```python
df = pd.read_csv ('police.csv', parse_dates=['stop_date'])
df['stop_year'] = df['stop_date'].dt.year
```

```python
# 1

# Option 1 - save the group in variable
g_year = df.groupby ('stop_year')
result = g_year['violation'].count()
result.sort_values(ascending=False)
```

```python
# Option 2 - method chaining, no need to save group in variable if its for quick calculation
df.groupby ('stop_year')['violation'].count().sort_values(ascending=False)
```

```python
# 2
df.groupby ('stop_year').agg({'violation_raw' : 'count',
                              'ticket_amount' : ['mean', 'sum']}).astype('int') #.transpose()
```

```python
# 3
# Preparinf the data
mask = df['stop_year'] == 2015
df_last = df[mask]
df_last.head(3)
```

```python
df_last.groupby ('driver_gender')['violation_raw'].count()
```

```python
# 4
df_last.groupby (['driver_gender','violation_raw']).agg({'driver_age' : 'mean', 
                                                         'ticket_amount' : ['count', 'sum'], 
                                                         'violation' : 'count'})
```
