```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Import data source

```python
# Import csv file
# Watch sample from the data
df = pd.read_csv ('police.csv', parse_dates=['stop_date'])
df['stop_year'] = df['stop_date'].dt.year
df.head(3)
```

## Basic visualization

```python
# How many violation for each gender

# df.groupby('driver_gender')['violation_raw'].count()

df.groupby('driver_gender')['violation_raw'].count().plot.pie()
```

```python
# autopct='%1.1f%%' >> The percentage of each part
# shadow=True >> Draw a small shadow under the pie chart
# explode=(0.1, 0) >> Move pieces out from the pie
# colors=('red','green') >> Touple that contains names of colors for the chart
# labels=('Female', 'Male') >> Touple that contains the labels of the series

df.groupby('driver_gender')['violation_raw'].count().plot.pie(autopct='%1.1f%%', shadow=True, explode=(0.1, 0),
                                                             colors=('red','green'), labels=('Female', 'Male'))
```

```python
# Notice that there are difference in the parameter names between the charts!


mask = df['stop_year'] == 2014
pt1 = df[mask].pivot_table(values='stop_year', index='violation_raw', columns='driver_gender', aggfunc='count')


pt1.plot.barh(width=0.8, color=('yellow','purple'), edgecolor='black')
```

```python
# Pre define the chart parameters:
# ylabel = label of y axis
# xlabel = label of x axis
# title = chart's title
# tick_params = controlling the ticks appearance
# xlim = the x axis ticks range

plt.ylabel('Violation Amount')
plt.xlabel('Year', fontweight='bold', color = 'green', fontsize='12', horizontalalignment='center')
plt.title ('Violation per Year', fontsize=17, color='blue')
plt.tick_params(axis='x', colors='red', direction='out', length=13, width=3)
plt.xlim(2005,2015)


df.groupby('stop_year')['violation_raw'].count().plot.line(color = 'magenta')
```
