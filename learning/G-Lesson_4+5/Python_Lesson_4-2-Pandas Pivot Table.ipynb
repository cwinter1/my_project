```python
import pandas as pd
import numpy as np
```

# Import data source

```python
# Import csv file
# Watch sample from the data
df = pd.read_csv ('police.csv', parse_dates=['stop_date'])
df['stop_year'] = df['stop_date'].dt.year
df.head(3)
```

## Pivot Tables

The function pivot_table() can be used to create spreadsheet-style pivot tables

* **values**: a column or a list of columns to aggregate

* **index**: a column to group by on the pivot table index

* **columns**: a column to group by on the pivot table column

* **aggfunc**: function to use for aggregation, defaulting to numpy.mean

```python
# For each gender, check how many violation were from each type

df.pivot_table(values='stop_year', index='violation_raw', columns='driver_gender', aggfunc='count')
```

```python
# Make the pivot table just on the last 3 years

mask = df['stop_year'] >= 2013
df[mask].pivot_table(values='stop_date', index='violation_raw', columns=['driver_gender', 'stop_year'], aggfunc='count')

```

```python
# 1
# Compare how many f/m were arrested after violation in the last 3 years

mask = df['stop_year'] >= 2013

df[mask].pivot_table(values='stop_date', index='is_arrested', columns=['stop_year', 'driver_gender'], aggfunc='count')
```

```python
# 2
# Add new column of decade by applying lambda function on all the df
df['age_decade'] = df.driver_age.apply (lambda x : str (x)[:-3] + '0s')
df['age_decade']
```

```python
# Pivot table as rewuested
df.pivot_table(values='stop_date', index='age_decade', columns='driver_gender', aggfunc='count')
```
