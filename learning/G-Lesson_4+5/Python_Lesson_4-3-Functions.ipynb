## Local and Global variables

```python
# variable scope - difference between local and global variable
# when x is out of the function - its global variable
# when x is defined in a function - its local variable(even if we have global variable with the same name)

def example1():
    x = 5
    print(x)
```

```python
# Calling the function

example1()
```

```python
# x is a local variable that exist only in the example1 function scope
# When we try to call the x variable, its not exist outside the function, so we get error
x
```

```python
y = 1      # global variable

def example1(y):
    print(y)
    
def example2():
    print(y)
```

```python
example1(10)

example2()
```

## Lambda Function

Lambda function is anonymous function which involve small and concise syntax

```python
# Regular user defined function

def identity_f(x):
    return x + 10
```

```python
# lambda function

lambda a : a + 10

# identity_l = lambda a : a + 10
```

```python
# Same result for both functions

identity_f(10)
# identity_l(10)
```

```python
# Lambda function can get more than one input

my_func = lambda a, b, c : a + b + c
```

```python
my_func(1, 2, 3)
```

## Combination of a Lambda and standart function 

#### First we see the example

```python
# A really cool use case of lambda expressions is within functions
# This function returns lambda expression

def my_func(n):
    return lambda a : a * n
```

```python
my_doubler = my_func(2)
my_tripler = my_func(3)
```

```python
my_doubler(11)
```

```python
my_tripler(11)
```

1.  Create a function that implement the string method "split":
    * The function signature is:
      def split(string,by):
       
    * Without using the split method str.split()
    
    * You can use: data types, data structures, conditions, loops, functions.
      Use as many variables as you like to the function
    
    * The function should return a value that is equal to the value that string.split() would return

```python
# Option 1
def split(string, by=' '):
    
    values = ['']
        
    for letter in string:
        
        if letter != by:
            values[-1] += letter
        else:
            values.append('')
    
    if values[-1] == '':
        values.pop()
    
    return values
```

```python
# Option 2
def split(string, by=''):
    result = []
    word = ''
    
    for letter in string:
        if letter == by:
            if word != '':
                result.append(word)
            word = ''
        else:
            word += letter
    
    if word != '':
        result.append(word) 
        
    return result
            
```

2. Write a Python function that takes a sequence of numbers and return True if all numbers are different from each other, otherwise return false

```python
def all_digits_unique(numbers):
    # `set` will only record unique elements.
    # We convert to a string and check if the unique
    # elements are the same number as the original elements.
    
    len_list = len(numbers)
    len_set = len( set(numbers) )
    return len_list == len_set



```

```python
test1 = [1,2,1,2,3,4,5,6]
all_digits_unique(test1)
```

```python
test2 =(1,2,3,4)
all_digits_unique(test2)
```
