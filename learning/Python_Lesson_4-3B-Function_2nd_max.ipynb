```python
import numpy as np
import pandas as pd
```

#### Create sample df

```python
df = pd.DataFrame(np.random.randint (1,11,(1000000,6)), 
                  columns=['a','b','c','d','e','f'])
df
```

#### Test the logic on one row

```python
sdf = list (df.loc[9])
sdf.sort()
sdf
```

```python
sdf[-2]
```

#### Put the logic in function

```python
def get_2nd_max (row):
    sdf = list (row)
    sdf.sort()
    return sdf[-2]
```

#### Test the function performance

```python
df['calc_2nd_max'] = df.apply(get_2nd_max, axis=1)
df
```
